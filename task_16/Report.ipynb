```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv("balance.txt", delim_whitespace=True)
```

```python
# Check the dataframe has loaded correctly
df
```

Write the code needed to produce a report that provides the following information:
* Compare the average income based on ethnicity.

```python
av_income_eth = df[["Income", "Ethnicity"]].groupby(["Ethnicity"]).mean()
av_income_eth
```

```python
print(f"The average income based on ethnicity is as follows, with incomes in the African American group slightly higher than the Asian or Caucasion groups:\n\n{av_income_eth.round(2)}")
```


* On average, do married or single people have a higher balance?

```python
bal_mar = df[["Balance", "Married"]].groupby(["Married"]).mean()
bal_mar
```

```python
print(f"On average sigle people have a very slightly higher balance of {bal_mar.iloc[0,0].round(2)} compared with married people with a balance of {bal_mar.iloc[1,0].round(2)}.")
```


* What is the highest income in our dataset?


```python
income_max = df["Income"].max()
income_max
```

```python
print(f"The highest income in the dataset is {income_max.round(2)}.")
```


* What is the lowest income in our dataset?

```python
income_min = df["Income"].min()
income_min
```

```python
print(f"The lowest income in the dataset is {income_min.round(2)}.")
```


* How many cards do we have recorded in our dataset? (Hint: use `sum()`)


```python
num_cards = df["Cards"].sum()
num_cards
```

```python
print(f"The total number of cards in the dataset is {num_cards}.")
```

* How many females do we have information for vs how many males? (Hint: use `count()` for a list of all methods for computation of descriptive stats, explore the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)).



```python
count_female = df["Gender"][df["Gender"] == "Female"].count()
count_female
```

```python
count_male = df["Gender"][df["Gender"] == "Male"].count()
count_male
```

```python
print(f"The total number of femails there is information for in the dataset is {count_female} and the total number of males there is information for in the dataset is {count_male}.")
```
